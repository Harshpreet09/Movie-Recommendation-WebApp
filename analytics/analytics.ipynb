```python
import pandas as pd

netflix = pd.read_csv("C:/Users/Administrator/Desktop/Project/Python/movie_recommendation_app/data/NetflixOriginals.csv")
prime = pd.read_csv("C:/Users/Administrator/Desktop/Project/Python/movie_recommendation_app/data/Amazon Prime Movies.csv")
imdb = pd.read_csv("C:/Users/Administrator/Desktop/Project/Python/movie_recommendation_app/data/IMDb Top 1000.csv")

# Preview
netflix.head()
prime.head()
imdb.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# IMDb genre plot
genre_counts = imdb['Genre'].str.split(',').explode().value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='viridis')
plt.title("Top 10 Genres in IMDb Top 1000")
plt.xlabel("Count")
```

```python
combined = pd.concat([
    netflix.assign(platform='Netflix'),
    prime.assign(platform='Amazon Prime'),
    imdb.assign(platform='IMDb')
], ignore_index=True)

combined['imdb_rating'] = combined['IMDB_Rating'].fillna(combined['IMDB_Rating'].mean())
combined['year'] = pd.to_numeric(combined['Released_Year'].fillna(0), errors='coerce')

sns.boxplot(x='platform', y='imdb_rating', data=combined)
plt.title("IMDb Rating Distribution by Platform")

```

```python
imdb['Released_Year'] = pd.to_numeric(imdb['Released_Year'], errors='coerce')
yearly = imdb['Released_Year'].value_counts().sort_index()

plt.plot(yearly.index, yearly.values)
plt.title("IMDb Movies by Release Year")
plt.xlabel("Year")
plt.ylabel("Count")
```

```python
##  Key Insights

- IMDb has more Action and Drama titles than Comedy.
- Amazon Prime movies are less rated compared to IMDb.
- Most titles are clustered between 2000 and 2022.
```
